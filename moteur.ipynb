```python
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer

path_data = "/home/lucien/datascience.stackexchange.com/Posts.xml"

posts = pd.read_xml((path_data), parser="etree", encoding="utf8")

def remove_html_tags(input_string):
    html_tags_pattern = re.compile('<.*?>')
    text_without_tags = re.sub(html_tags_pattern, '', input_string)
    return text_without_tags

posts['body_tokenized'] = None

# Créez un objet CountVectorizer pour effectuer la tokenization
vectorizer = CountVectorizer()

for i, content in enumerate(posts['Body']):
    try:
        tokens = vectorizer.build_tokenizer()(remove_html_tags(content))
        posts.at[i, 'body_tokenized'] = [tokens]
    except:
        pass

#on supprime les lignes où le Body vaut None

posts.head(10)
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
```

```python
posts['body_tokenized_stop_words'] = None
# Obtenez les stop words en anglais
stop_words = set(stopwords.words('english'))

#on veut une foncion qui remplace les caractères majuscules par des minuscules

def lower_case(input_string):
    return input_string.lower()

#on boucle sur la colonne body_tokenized pour supprimer les stop words

for i, content in enumerate(list(posts['body_tokenized'])):
    if content is not None:
        posts['body_tokenized_stop_words'][i] = [lower_case(word) for word in content[0] if not word in stop_words]
        
```

```python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()
posts['body_tokenized_stop_words_lemma'] = None
```

```python
# pour l'indexation le mieux serait le dictionnaire
import pandas as pd
vocab = []
voc = []
for index, lemma in enumerate(posts["body_tokenized_stop_words_lemma"]): #lemma liste de mot d'un post
    #on veut récupérer la valeur de la colonne id à cette ligne
    article = posts.loc[index, 'Id']

    if lemma is not None:
        for word in lemma:#voc contient tous les mots de tous les posts
            voc.append(word) 
            vocab.append((word, article))
    
print(vocab[:100])

vocab_sans_repet = list(set(voc))
print(vocab_sans_repet[:100])

index_inv = {}
for voca in vocab_sans_repet:
    index_inv[voca] = []

print("vocab la liste de tuple:", vocab[:100])

#on parcoure vocab pour la séparer en liste de liste de tuple lorsque l'élément droit du tuple change
liste_vocab = []
liste = []
for i, voca in enumerate(vocab):
    if i == 0:
        liste.append(voca)
    else:
        if voca[1] == vocab[i-1][1]:
            liste.append(voca)
        else:
            liste_vocab.append(liste)
            liste = []
            liste.append(voca)
print(liste_vocab[:10])
```

```python
index_inv_occ = {}
for voca in vocab_sans_repet:
    index_inv_occ[voca] = []

i = 0
for voca in liste_vocab:
    i += 1
    for tuple in voca:
        # print(voca[0][0])
        # print(voca.count(voca[0]))
        # break
        index_inv_occ[tuple[0]].append((tuple[1], voca.count(tuple)))

print(index_inv_occ['avoid'])

#créons un dictionnaire qui pour chaque mots donne son nombre d'occurence dans le corpus

dico = {}
for mot in voc:
    if mot in dico:
        dico[mot] += 1
    else:
        dico[mot] = 1
```

Mettons en place notre méthode recherche

```python
def recherche_proba(query, index_inv_occ, documents):
    scores = {}
    
    # Calculer le score pour chaque document
    for keyword in query:
        if keyword in index_inv_occ:
            for doc_id, occ in index_inv_occ[keyword]:
                if doc_id not in scores:
                    scores[doc_id] = 0
                scores[doc_id] += occ / dico[keyword] #/ documents[doc_id] #on favorise les mots qui ont une grosse pondération par rapport 
    
    # # Normaliser les scores par la longueur des documents
    # for doc_id in scores:
    #     scores[doc_id] /= len(documents[doc_id])

    
    # Trier les documents par score décroissant
    sorted_scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    
    return sorted_scores
```

```python
#Créons une fonction de transformation de la requête en liste de mots, utilisant la même méthode qu eprécédemment


def query_to_list(query):
    L = []
    vectorizer = CountVectorizer()
    tokens = vectorizer.build_tokenizer()(remove_html_tags(query[0]))
    query = [lower_case(word) for word in tokens if not word in stop_words]
    lemmatizer = WordNetLemmatizer()
    for mot in query:
        L.append(lemmatizer.lemmatize(mot))
    return L
```

```python
documents = {}
i = 0
for index in posts['Id']:
    
    if posts['body_tokenized_stop_words_lemma'][i] is not None:
        documents[index] = len(posts['body_tokenized_stop_words_lemma'][i])
    i += 1

print(documents)
```

```python
query = ['I would like to learn how neural networks work.']
query = query_to_list(query) 

print(recherche_proba(query, index_inv_occ, documents)[:50])
```
